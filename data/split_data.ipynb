```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../clean_adhd.csv")
```

```python
set(df['dx'])
```

```python
use_df = df[df['dx']!='pending'].reset_index()
```

```python
use_df
```

```python
names = np.array(use_df['new_name'])
sites = np.array(use_df['site'])
labels = np.array([0 if f == 0 else 1 for f in use_df['dx'] ])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
for train_index, test_index in sss.split(names, sites):
    train_names = names[train_index]
```

```python
is_train = []
for name in use_df['new_name']:
    if name in train_names:
        is_train.append(1)
    else:
        is_train.append(0)
```

```python
use_df['is_train'] = is_train
```

```python
use_df
```

```python
use_df['new_dx'] = [0 if f == '0' else '1' for f in use_df['dx'] ]
```

```python
use_df.to_csv('use_adhd.csv',index=None)
```
